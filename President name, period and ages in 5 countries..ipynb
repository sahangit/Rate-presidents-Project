```python
# import libraries

import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_presidents = pd.read_csv('President datasheet 5 countries.csv')
```

```python
df_growth = pd.read_csv('GDP per Capita Growth Rate.csv')
```

### Growth data

```python
# rename columns


df_growth.columns = ['Date','China','United Kingdom', 'India', 'Sri Lanka', 'Singapore', 'United States']
df_growth.tail()
```

```python
# check type

for column in range(df_growth.shape[1]):
    typetest = []
    for row in range(df_growth.shape[0]):
        tp = type(df_growth.iloc[:,column][row])
        if tp not in typetest:
            typetest.append(tp)
    print('type of '+str(df_growth.columns[column])+' is '+str(typetest))
        
        
```

### President Data

```python
df_presidents.head()
```

```python
# let us derive in another data frame the country, President and Political reign
```

```python
# creating new dataframe from
df_reigns = df_presidents.iloc[:,]
df_reigns['Experience'] = df_reigns['Year start']-df_reigns['First political exp']
df_reigns.drop('First political exp', inplace = True,  axis = 1)

# adding columns
df_reigns[["Max",'Min','Average','Std Dev']] = ""
df_reigns.head()
```

```python
# max,min,average, std dev

for row in range(len(df_reigns)):
    country = df_reigns['Country'][row]
    start = df_reigns['Year start'][row]
    end = df_reigns['Year end'][row]
    
    
    try:
        index_start = df_growth[df_growth['Date'] == int(start)].index[0]
    
    except:
        index_start = 100        
        
    try:
        index_end = df_growth[df_growth['Date'] == int(end)].index[0]+1
    
    except:
        index_end = 100
    
    df_reigns['Max'][row] = df_growth[country].iloc[index_start:index_end].max()
    df_reigns['Min'][row] = df_growth[country].iloc[index_start:index_end].min()
    df_reigns['Average'][row] = df_growth[country].iloc[index_start:index_end].mean()
    df_reigns['Std Dev'][row] = df_growth[country].iloc[index_start:index_end].std()

```
