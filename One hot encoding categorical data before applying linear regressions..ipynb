# Binary Encoding presidential data

### importing libraries and datasets

```python
import pandas as pd 
```

```python
df_presidents = pd.read_csv('Growth mapped to reigns.csv')
pd.set_option('display.max_columns', None)
df_presidents.head()
```

```python
columns = ['Country','Presidential term','Year start','Educational Background','Professional Background',
           'Political Party','University','Left Hand / Right Handed']

for column in range(len(columns)):
    print(df_presidents.groupby(by = columns[column])['Name'].count()*100/df_presidents['Name'].count())
```

```python
# These items are removed from furthuer classifications as they cannot be used accross differenct countries
# or not enough unique items.
# 'end year','Uniersity'

columns = ['Country', 'Name','Presidential term','Year start', 'Educational Background', 'Professional Background',
       'Left Hand / Right Handed','Experience', 'Average']

df_presidents_final = df_presidents[columns]
df_presidents_final.shape
```

```python
# dropping null values in average column and re-categorizing educational background as 
# High School, Degree, and Higher Degree

df_presidents_final = df_presidents_final.dropna().reset_index()

for row in range(len(df_presidents_final)):
    entry = df_presidents_final['Educational Background'][row]
    
    if (entry == 'Bachelor’s' or entry == 'Degree' or entry == 'Law' or entry =='JD' ):
        df_presidents_final['Educational Background'][row] = 'Degree'
        
    elif (entry == 'Diploma(Phd followed)' or entry == 'Masters'):
        df_presidents_final['Educational Background'][row] = 'Higher Degree'
    
    elif (entry =='LLD' or entry == 'PhD' or entry == 'DPhil'):
        df_presidents_final['Educational Background'][row] = 'Doctarate'
    
    else:
        df_presidents_final['Educational Background'][row] = 'No Degree'
```

```python
# Let us also reduce number of categories for Professional Backgrounds

for row in range(len(df_presidents_final)):
    entry = df_presidents_final['Professional Background'][row]
    
    if entry == 'Politician':
        df_presidents_final['Professional Background'][row] = 'Politician'
    else:
        df_presidents_final['Professional Background'][row] = 'Non-Politician'
```

```python
df_presidents_final.head(10)
```

```python
df_presidents_final['Professional Background'].unique()
```

```python
columns = df_presidents_final.columns[3:-2].tolist()

for column in range(len(columns)):
    print(df_presidents_final.groupby(by = columns[column])['Name'].count()*100/df_presidents['Name'].count())
```

```python
# In here country, Professional background and Handedness are nominal variables
# Educational Backgroud is ordinal

# Let us encode Educational Background according Sri Lanka Qualification framework
# No Degree = 0, Degree = 5, Phd = 10, So higher degree = 7.5

df_presidents_final_map = df_presidents_final
df_presidents_final_map['Educational Background'] = df_presidents_final['Educational Background'].map({'No Degree':0, 'Degree':5, 'Higher Degree':7.5, 'Doctarate':10})
df_presidents_final_map.head()
```

```python
# encoding Left = 0, Right = 1
# encodin Politicial = 0, Non-Politician = 1

df_presidents_final_map['Professional Background'] = df_presidents_final['Professional Background'].map({'Politician':0, 'Non-Politician':1,})
df_presidents_final_map['Left Hand / Right Handed'] = df_presidents_final['Left Hand / Right Handed'].map({'Left':0, 'Right':1,})
df_presidents_final_map
```

```python
# one hot encoding for the country.
df_dummy = pd.get_dummies(df_presidents_final_map['Country'], prefix='Country')
```

```python
df_presidents_final_map = pd.concat([df_dummy, df_presidents_final_map], axis = 1)
df_presidents_final_map['index'] = df_presidents_final_map.index

df_presidents_final_map.drop(columns = ['Country','Name'], inplace = True)
df_presidents_final_map
```

```python
df_presidents_final_map.to_csv('President and Gdp data Encoded.csv', index=False)
```
